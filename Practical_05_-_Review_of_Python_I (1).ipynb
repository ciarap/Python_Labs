#### Question 3: Function returning a list 

__(a)__ Write a python function

    def f(n):

where _n_ is a positive integer, that returns a list of all non-negative integers which are divisible by 7 but are not a multiple of 5 up to but not including _n_.

```python
# Answer 3
def f(n):
    return [k for k in range(n) if k%7==0 and k%5!=0]
f(50)
```

#### Question 4: function returning a dict 

Write a python function

    def wordLength1(data):

where _data_ is a list of strings, and returns a dict with key-value pairs (string $\to$ length of string). For exaple 

    wordLength1(['This', 'sentence', 'is','false'])

should return

    {'This': 4, 'false': 5, 'is': 2, 'sentence': 8}

Hint: use the _len_ function to get the length of a string.

```python
# Answer 4
def wordLength1(data):
    return {word:len(word) for word in data}
wordLength1(['This', 'sentence', 'is','false'])
```

#### Question 5: function returning a dict 

Write a python function

    def wordLength2(string):

where _data_ is a string of words, and returns a dict with key-value pairs (string $\to$ length of string) for each word in the string. For exaple 

    wordLength2(['This sentence is false'])

should return

    {'This': 4, 'false': 5, 'is': 2, 'sentence': 8}

Hint: use the string _split_ method.



```python
# Answer 5
def wordLength2(string):
    return {word:len(word) for word in string.split()}
wordLength2("This sentence is false")
    
```

#### Question 6: function calculating a recursive sequence

Write a python function

    def f(n):
    
where _n_ is a positive integer, and returns the n(th) Fibonacci, where the Fibonacci numbers are defined by

\\[
    f_n = \begin{cases} 1  & n=1\\1  & n=2\\ f_{x-1}+f_{x-2} & n>2\end{cases}
\\]

```python
# Answer 6
def f(n): 
    if n==1 or n==2:
        return 1
    else :
        return  f(n-1) + f(n-2)

f(12)
```

Write a python function 

    def f():

which returns a list of all positive integers between _1000_ and _3000_ (including _1000_ but not including _3000_) such that each digit of the number is an even number. 

Hint: Use function _str_ to convert an integer to a string.
