## A Human-Vampire Interaction  Models

If we define

 * $h(t)$ to be the human population at time $t$
 * $v(t)$ to be the vampire population at time $t$

then the population at time $t+1$ is given by

\\[
    \begin{array}{rcl}
    h(t+1) &=& a \cdot h(t) - c \cdot h(t)\cdot v(t)\\
    v(h+1) &=& (b_1-b_2) \cdot h(t)\cdot v(t) - d \cdot v(t)
    \end{array}
\\]

where parameters

 * $a$ is the growth rate of the human population (in the absence of vampires)
 * $c$ is the probability of a human dying in a human-vampire interaction.
 * $b_1$ is the probability of a human being turned to a vampire in a human-vampire interaction.
 * $b_2$ is the probability of a vampire dying in a human-vampire interaction.
\\[
    b = b_1-b_2
\\]
 * $d$ is the death rate of vampire by staking due to vampire hunters.

### The Stoker-King model

 * The vampire bites the victim and drinks the victim’s blood, then returns to feed for 4-5 consecutive days, whereupon the victim dies, is buried and rises to become another vampire (unless a wooden stake is put through its heart). In both stories, very few were staked, so
\\[
    b_1 = 0.99
\\]
 * Vampires usually need to feed every day, so more and more human beings are constantly turned into vampires, so time needs to count in days not years.
 * Initial conditions of the Stoker-King model are the following: 1 vampire, there are no organized groups of vampire slayers. 
\\[
    c = 1, \quad d = 0
\\]
 * The slayers, Jonathan Harker and Abraham van Helsing, were not effective, so
\\[
    b_2 = 0
\\]
 * Initial conditions: Lets pick an arbitray date, say 1 January 2011, and the human population is 7 Billion, and lets say there is one (hungry) vampire ...

```python
h,v = [],[]
a= 1.75E3**(1/10011)
b1,b2= 0.99, 0
c=1
d=0

h,v = [],[]

for k in range(10):
    hNow = h[-1]
    vNow = v[-1]
    h.append(a*hNow - c*hNow*vNow)
    v.append((b1-b2)*hNow - d*vNow)
```
